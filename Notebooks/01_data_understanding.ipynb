## load raw weather data and labels

```python
# Load main weather dataset
df = pd.read_csv('../Data/Supervised/weather_prediction_dataset.csv')

# Load the labels (target variable)
labels = pd.read_csv('../Data/Supervised/weather_prediction_picnic_labels.csv')

print("Data loaded successfully")
```

## check for missing values

```python
# Check for missing values
print(df.isnull().sum().sum())
print(f"Total missing values: {df.isnull().sum().sum()}")

# Show missing values per column (only columns that have them)
missing = df.isnull().sum()
print(missing[missing > 0])
```

## analyse class balance across all 18 stations

```python
labels.head()
```

```python
# Check the balance of our target variable
# How many True vs False days per station?
print(labels.describe())
print("\nValue counts for Basel label:")
print(labels['BASEL_picnic_weather'].value_counts())

print("\nPercentage count  for Basel label:")
print(labels['BASEL_picnic_weather'].value_counts(normalize=True))
```

```python
# Check True/False balance across ALL stations
label_cols = [col for col in labels.columns if col != 'DATE']

balance = labels[label_cols].mean()
print("Proportion of 'True' (favorable) days per station:")
print(balance.sort_values())
```

## merge datasets on DATE

```python
print("Weather data DATE sample: ", df['DATE'].head().tolist())
```

```python
print("Labels DATE sample:", labels['DATE'].head().tolist())
```

```python
print("\nDate column types:")
print("Weather:", df['DATE'].dtype)
print("Labels:", labels['DATE'].dtype)
```

```python
# Merge weather data with labels on the DATE column
df_merged = pd.merge(df, labels, on='DATE', how='inner')

print(f"Original weather data shape: {df.shape}")
print(f"Labels shape: {labels.shape}")
print(f"Merged dataframe shape: {df_merged.shape}")
print(f"\nMerge successful: {len(df_merged) == len(df)}")
```

## convert DATE to proper datetime format

```python
# Convert DATE from integer (20000101) to proper datetime
df_merged['DATE'] = pd.to_datetime(df_merged['DATE'], format='%Y%m%d')

print("Date conversion successful")
print(f"Date range: {df_merged['DATE'].min()} to {df_merged['DATE'].max()}")
print(f"Total days: {len(df_merged)}")
```

## save analysis-ready dataset

```python
# Save the merged dataset so we don't need to repeat the merge process in future notebooks

df_merged.to_csv('../Data/Supervised/weather_merged.csv', index=False)
print("Merged dataset saved successfully")
```
